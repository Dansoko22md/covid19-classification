# Requêtes Graph-RAG

```python
import pandas as pd
import numpy as np
from pathlib import Path
from step5_rag import GraphRAGSimplified
```

```python
BASE_PATH = Path("../data/processed")
rag = GraphRAGSimplified(BASE_PATH / "article_graph.gpickle", BASE_PATH / "embeddings.npy", BASE_PATH / "articles_with_communities.csv")
```

```python
QUERY = "COVID-19 vaccine efficacy"
METHOD = "hybrid"
TOP_K = 5
ans = rag.answer_question(QUERY, method=METHOD, top_k=TOP_K)
pd.DataFrame(ans['results'])[['score','title','community']]
```

```python
print(ans['context'])
ans['communities_covered']
```

```python
ans_comm = rag.answer_question("vaccine adverse effects", method="community", top_k=5)
pd.DataFrame(ans_comm['results'])[['score','title','community']]
```

```python
INDEX = 0
sim = rag.find_similar_articles(INDEX, top_k=5, method="hybrid")
pd.DataFrame(sim)[['score','title']]
```

```python
import matplotlib.pyplot as plt
df_res = pd.DataFrame(ans['results'])
if 'community' in df_res.columns:
    c = df_res['community'].dropna()
    s = c.value_counts()
    plt.figure(figsize=(8,4))
    plt.bar(range(len(s)), s.values)
    plt.xticks(range(len(s)), s.index.astype(int))
    plt.title('Couverture des communautés')
    plt.tight_layout()
    plt.show()
```

```python
out = BASE_PATH / 'rag_last_results.csv'
pd.DataFrame(ans['results']).to_csv(out, index=False)
str(out)
```
